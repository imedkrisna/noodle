```python
import pandas as pd
import world_trade_data as wits
import pandas as pd
import datetime
pd.set_option('display.max_rows', 999)
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
pd.options.display.max_rows=999
# REMOVE THIS WHEN NOT MAKING PLOTS!!!!
jtplot.reset()
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", context='paper',font_scale=1,rc=custom_params)
```

```python
b=wits.get_products()
b=b.loc[(b.grouptype=='Sector') & (
    b.index!='Total')].productdescription.values
z=["01-05_Animal","06-15_Vegetable","16-24_FoodProd"]
```

```python
z="01-05_Animal"
a=wits.get_indicator(indicator='RCA',
                     reporter='idn',year=['2002','2021'],product=z)
a=a.reset_index()
b=wits.get_indicator(indicator='RCA',
                     reporter='vnm',year=['2002','2021'],product=z)
b=b.reset_index()
c=wits.get_indicator(indicator='RCA',
                     reporter='mys',year=['2002','2021'],product=z)
c=c.reset_index()
d=wits.get_indicator(indicator='RCA',
                     reporter='tha',year=['2002','2021'],product=z)
d=d.reset_index()
e=wits.get_indicator(indicator='RCA',
                     reporter='sgp',year=['2002','2021'],product=z)
e=e.reset_index()
wew=pd.concat([a,b,c,d,e]).reset_index()
```

```python
#sns.set_theme(context='paper',style='white',font_scale=1.5)
g=sns.lineplot(data=wew,x='Year',y='Value',hue='Reporter',linewidth=2.5,style='Reporter')
g.set_xticks(['2002','2005','2008','2011','2014','2017'])
g.legend(bbox_to_anchor=(1,-0.15),ncol=3)
plt.ylim([0, 4.4])
plt.yticks([0,1,2,3,4])
plt.ylabel('')
plt.title("RCA on animal products of 5 ASEAN countries\n") # Change this
plt.savefig('RCA_Animal.png',transparent=False,dpi=100,bbox_inches = 'tight') # Change this
```
